# Grafico 1: Cantidad de operaciones vs Tama√±o

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Obtiene el directorio actual
directorio_actual = os.getcwd()

# Nombre del archivo CSV que se desea abrir
nombre_archivo = "treesort_resultados.csv"

# Se Construye la ruta completa del archivo utilizando os.path.join
ruta_completa = os.path.join(directorio_actual, nombre_archivo)

# Verifico si el archivo existe antes de intentar abrirlo
if os.path.exists(ruta_completa):
    # Abre el archivo CSV utilizando pandas
    df = pd.read_csv(ruta_completa)

else:
    print(f"El archivo {nombre_archivo} no existe en el directorio actual.")
```

```python
# Crear gr√°fico
plt.figure(figsize=(10, 6))
plt.plot(df["Tama√±o"], df["Operaciones"], color='b', label="Operaciones")

# Personalizaci√≥n del gr√°fico
plt.title("Cantidad de Operaciones vs Tama√±o de Entrada", fontsize=14)
plt.xlabel("Tama√±o de Entrada", fontsize=12)
plt.ylabel("Cantidad de Operaciones", fontsize=12)
plt.grid(True, linestyle='-', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar gr√°fico
plt.show()
```
