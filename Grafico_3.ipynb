# Grafico 3: Comparación experimental vs teórica

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Obtiene el directorio actual
directorio_actual = os.getcwd()

# Nombre del archivo CSV que se desea abrir
nombre_archivo = "treesort_resultados.csv"

# Se Construye la ruta completa del archivo utilizando os.path.join
ruta_completa = os.path.join(directorio_actual, nombre_archivo)

# Verifico si el archivo existe antes de intentar abrirlo
if os.path.exists(ruta_completa):
    # Abre el archivo CSV utilizando pandas
    df = pd.read_csv(ruta_completa)

else:
    print(f"El archivo {nombre_archivo} no existe en el directorio actual.")
```

```python
# Reemplazo el sufijo 's' a vacío para evitar inconvenientes y convierto a float
df['Tiempo(ms)'] = df['Tiempo(ms)'].replace(r's$', '', regex=True).astype(float)

# Generar valores teóricos basados en O(n log n)
df['n_log_n'] = df['Tamaño'] * np.log2(df['Tamaño'])

# Escalar los valores teóricos para compararlos con los experimentales
factor_ajuste = max(df['Tiempo(ms)']) / max(df['n_log_n'])
df['n_log_n'] *= factor_ajuste

# Crear gráfico de comparación
plt.figure(figsize=(10, 6))

# Gráfico experimental
plt.plot(df['Tamaño'], df['Tiempo(ms)'], label='Tiempo de Ejecución Experimental', color='b')

# Gráfico teórico
plt.plot(df['Tamaño'], df['n_log_n'], label='Comportamiento Teórico O(n log n)', color='r', linestyle='-')

# Personalización del gráfico
plt.title("Comparación: Tiempo de Ejecución vs Tamaño de", fontsize=14)
plt.xlabel("Tamaño de Entrada", fontsize=12)
plt.ylabel("Tiempo de Ejecución (ms)", fontsize=12)
plt.grid(True, linestyle='-', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

# Comparación: Mejor, Promedio y Peor Caso (Tiempo de Ejecución)

```python
# Reemplazo el sufijo 's' a vacío y convierto a float
df['Tiempo(ms)'] = df['Tiempo(ms)'].replace(r's$', '', regex=True).astype(float)

# Generar valores teóricos para O(n log n) (mejor caso)
df['n_log_n'] = df['Tamaño'] * np.log2(df['Tamaño'])

# Generar valores teóricos para O(n^2) (peor caso)
df['n_squared'] = df['Tamaño'] ** 2

# Escalar los valores teóricos para compararlos con los experimentales
factor_ajuste_log = max(df['Tiempo(ms)']) / max(df['n_log_n'])
factor_ajuste_sq = max(df['Tiempo(ms)']) / max(df['n_squared'])
df['n_log_n'] *= factor_ajuste_log
df['n_squared'] *= factor_ajuste_sq

# Crear gráfico de comparación
plt.figure(figsize=(10, 6))

# Gráfico experimental
plt.plot(df['Tamaño'], df['Tiempo(ms)'], label='Tiempo de Ejecución Experimental', color='b')

# Gráfico teórico para O(n log n)
plt.plot(df['Tamaño'], df['n_log_n'], label='Mejor Caso O(n log n)', color='g', linestyle='-')

# Gráfico teórico para O(n^2)
plt.plot(df['Tamaño'], df['n_squared'], label='Peor Caso O(n^2)', color='r', linestyle='-')

# Personalización del gráfico
plt.title("Comparación: Mejor, Promedio y Peor Caso (Tiempo de Ejecución)", fontsize=14)
plt.xlabel("Tamaño de Entrada", fontsize=12)
plt.ylabel("Tiempo de Ejecución (ms)", fontsize=12)
plt.grid(True, linestyle='-', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```
